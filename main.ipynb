## Genres

```python
cnt = {}
for genres in data['genres']:
    for genre in genres.split('|'):
        cnt[genre] = cnt.get(genre, 0) + 1
cnt
```

```python
genres = []
for title_genres in data['genres']:
    for genre in title_genres.split('|'):
        if genre not in genres:
            genres.append(genre)
genres
```

```python
is_genre_columns = data['genres'].dropna().to_frame()
def fill_genre(row, genre):
    if genre in row['genres'].split('|'):
        return 1
    else:
        return 0
for genre in genres:
    is_genre_columns[genre] = is_genre_columns.apply(fill_genre, axis=1, genre=genre)
is_genre_columns = is_genre_columns.drop(columns='genres')
is_genre_columns
```

## plot_keywords

```python
cnt = {}
for keywords in data['plot_keywords'].dropna():
    for keyword in keywords.split('|'):
        cnt[keyword] = cnt.get(keyword, 0) + 1
```

```python
plot_keywords = []
for keywords in data['plot_keywords'].dropna():
    for keyword in keywords.split('|'):
        if keyword not in plot_keywords and cnt[keyword] > 10:
            plot_keywords.append(keyword)
len(plot_keywords)
```

```python
keywords_columns = data['plot_keywords'].drop(index=rows_to_delete).dropna().to_frame()
def fill_keyword(row, keyword):
    if keyword in row['plot_keywords'].split('|'):
        return 1
    else:
        return 0
for keyword in plot_keywords:
    keywords_columns[keyword] = keywords_columns.apply(fill_keyword, axis=1, keyword=keyword)
keywords_columns = keywords_columns.drop(columns='plot_keywords')
keywords_columns
```

# Recomendation 

```python
indexes = []
for index in is_genre_columns.index:
    if index % 1000 == 0:
        print(index)
    if index in keywords_columns.dropna().index and index in content_rating_column.dropna().index:
        indexes.append(index)
```

```python
X = pd.concat([is_genre_columns.loc[indexes].drop(columns=['Film-Noir', 'Reality-TV', 'Short', 'News']), keywords_columns.loc[indexes],
                                                  content_rating_column.loc[indexes]], axis=1)
X
```

```python
cosine_sim = cosine_similarity(X, X)
cosine_sim
```

```python
def get_recommendations(index, cosine_sim=cosine_sim):
    
    sim_scores = list(enumerate(cosine_sim[index]))
    
    # Sort the movies based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # Get the scores of the 5 most similar movies
    sim_scores = sim_scores[0:6]
    
    # Get the movie indices
    movie_indices = [i[0] for i in sim_scores]
    
    # Return the top 5 most similar movies
    return data.loc[X.iloc[movie_indices].index]
```
